```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
url = "https://github.com/ageron/data/raw/main/lifesat/lifesat.csv"
df = pd.read_csv("https://github.com/ageron/data/raw/main/lifesat/lifesat.csv")
df.head()
```

```python
df.info()
```

```python
X = df[["GDP per capita (USD)"]].values
y = df[["Life satisfaction"]].values
```

```python
df.plot(kind = "scatter", x="GDP per capita (USD)", y="Life satisfaction")
plt.plot()
```

```python
model1 = LinearRegression()
model1.fit(X,y)
print(f"The models predicted value is: {model1.predict([[33442.8]])}")
```

```python
from sklearn.neighbors import KNeighborsRegressor
model2 = KNeighborsRegressor(n_neighbors=1)
model3 = KNeighborsRegressor(n_neighbors=3)
model2.fit(X,y)
print(f"model 1: {model2.predict([[33442.8]])}")
model3.fit(X,y)
print(f"model 2: {model3.predict([[33442.8]])}")
```
